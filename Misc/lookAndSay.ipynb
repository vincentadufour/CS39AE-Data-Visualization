# Look-And-Say Sequence Graphs
The below is an attempt to find different ways to graph the Look-And-Say sequence to n iterations.

```python
import matplotlib.pyplot as plt

# look-and-say sequence
def look_and_say(n):
    sequence = "1"
    for _ in range(n - 1):
        next_sequence = ""
        i = 0
        while i < len(sequence):
            count = 1
            while i + 1 < len(sequence) and sequence[i] == sequence[i + 1]:
                i += 1
                count += 1
            next_sequence += str(count) + sequence[i]
            i += 1
        sequence = next_sequence
    return sequence

# count the frequency of each number
def count_digits(sequence):
    counts = {"1": 0, "2": 0, "3": 0}
    for digit in sequence:
        if digit in counts:
            counts[digit] += 1
    return counts

# plot using matplotLib
def plot_counts(counts):
    labels = list(counts.keys())
    values = list(counts.values())

    bars = plt.bar(labels, values, color=['blue', 'orange', 'green'])

    for bar in bars: 
      plt.text( 
        bar.get_x() + bar.get_width() / 2, # X position: center of each bar 
        bar.get_height(), # Y position: height of each bar 
        bar.get_height(), # Text to display (the height of the bar) 
        ha='center', # Horizontal alignment 
        va='bottom' # Vertical alignment 
      )
    
    plt.xlabel('Digit')
    plt.ylabel('Frequency')
    plt.title('Frequency of Digits in Look-And-Say Sequence')
    plt.show()

    

def look_and_say_analysis(iterations):
    sequence = look_and_say(iterations)
    counts = count_digits(sequence)
    plot_counts(counts)

# Example usage
look_and_say_analysis(10)
```

```python
import matplotlib.pyplot as plt

def look_and_say_line_graph(iterations):
    counts_over_time = {"1": [], "2": [], "3": []}
    
    sequence = "1"
    for _ in range(iterations):
        counts = count_digits(sequence)
        for digit in "123":
            counts_over_time[digit].append(counts[digit])
        
        # Generate the next sequence
        next_sequence = ""
        i = 0
        while i < len(sequence):
            count = 1
            while i + 1 < len(sequence) and sequence[i] == sequence[i + 1]:
                i += 1
                count += 1
            next_sequence += str(count) + sequence[i]
            i += 1
        sequence = next_sequence

    # Plot the line graph
    plt.plot(range(1, iterations + 1), counts_over_time["1"], label='Digit 1')
    plt.plot(range(1, iterations + 1), counts_over_time["2"], label='Digit 2')
    plt.plot(range(1, iterations + 1), counts_over_time["3"], label='Digit 3')

    plt.xlabel('Iteration')
    plt.ylabel('Frequency')
    plt.title('Frequency of Digits in Look-And-Say Sequence Over Time')
    plt.legend()
    plt.show()

# Example usage
look_and_say_line_graph(50)
```

```python
import matplotlib.pyplot as plt

def look_and_say_pie_chart(iteration):
    sequence = look_and_say(iteration)
    counts = count_digits(sequence)
    
    labels = ['1', '2', '3']
    sizes = [counts['1'], counts['2'], counts['3']]
    colors = ['blue', 'orange', 'green']

    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
    plt.title(f'Composition of Digits in Look-And-Say Sequence at Iteration {iteration}')
    plt.show()

# Example usage
look_and_say_pie_chart(50)
```
