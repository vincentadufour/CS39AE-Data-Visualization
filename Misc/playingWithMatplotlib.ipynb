```python
import matplotlib.pyplot as plt

# Step 2: Prepare data
x = [1, 2, 3, 4, 5]  # X-axis values
y = [2, 3, 5, 7, 11]  # Y-axis values

# Step 3: Create a scatterplot
plt.scatter(x, y)

# Calculate the best fit line
slope, intercept = np.polyfit(x, y, 1)  # '1' indicates a linear fit (first-degree polynomial)
trendline = np.array(x) * slope + intercept  # Compute y values of the trend line
print(slope, "...", intercept, "...", trendline)


# Plot the trend line
plt.plot(x, trendline, color='red', linestyle='--')  # Red dashed line for the trend line

# Step 4: Add titles and labels
plt.title('Simple Scatterplot Example')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Step 5: Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data list
data_list = [
    11, 11, 11, 13, 16, 20, 22, 23, 24, 24, 26, 26, 27, 27, 28, 30, 31, 31,
    35, 35, 35, 35, 35, 35, 36, 36, 36, 37, 37, 38, 38, 39, 39, 39, 39, 41,
    43, 43, 45, 45, 45, 48, 49
]

# Plot the histogram
plt.figure(figsize=(8, 6)) # axis size
count, bins, ignored = plt.hist(data_list, bins=range(10, 51, 5), alpha=0.6, color='g', edgecolor='black', density=True)
print(count)
print(bins)

# Calculate mean and standard deviation using Numpy functions
mu = np.mean(data_list)
std = np.std(data_list)

# Plot the PDF (Probability Density Function) using manual calculations
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / std) ** 2)

plt.plot(x, p, 'k', linewidth=2, label=f'Fit line (={mu:.2f}, ={std:.2f})')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Best Fit Line')
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate the five-number summary
minimum = np.min(data)
q1 = np.percentile(data, 25)
median = np.median(data)
q3 = np.percentile(data, 75)
maximum = np.max(data)

# Create a vertical box plot
plt.figure(figsize=(6, 8))
box = plt.boxplot(data, vert=True, patch_artist=True)

# Annotate the five-number summary on the plot
plt.text(1.1, minimum, f'Min: {minimum:.2f}', ha='left', va='center')
plt.text(1.1, q1, f'Q1: {q1:.2f}', ha='left', va='center')
plt.text(1.1, median, f'Median: {median:.2f}', ha='left', va='center')
plt.text(1.1, q3, f'Q3: {q3:.2f}', ha='left', va='center')
plt.text(1.1, maximum, f'Max: {maximum:.2f}', ha='left', va='center')

# Additional plot details
plt.title("Vertical Box Plot with Five-Number Summary Annotations")
plt.ylabel("Value")
plt.xticks([1], ['Data'])
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data
data = np.random.randn(100)
x = np.random.rand(100)
y = x + np.random.randn(100) * 0.1
z = np.linspace(0, 10, 100)
line_y = np.sin(z)

# Create a figure and a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot a histogram
axs[0, 0].hist(data, bins=20, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Histogram')
axs[0, 0].set_xlabel('Value')
axs[0, 0].set_ylabel('Frequency')

# Plot a scatterplot
axs[0, 1].scatter(x, y, color='green', alpha=0.5)
axs[0, 1].set_title('Scatterplot')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')

# Plot a box plot
axs[1, 0].boxplot(data)
axs[1, 0].set_title('Box Plot')
axs[1, 0].set_ylabel('Value')

# Plot a line chart
axs[1, 1].plot(z, line_y, color='blue', linestyle='-', linewidth=2)
axs[1, 1].set_title('Line Chart')
axs[1, 1].set_xlabel('Z')
axs[1, 1].set_ylabel('sin(Z)')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```
