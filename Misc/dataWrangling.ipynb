```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd 

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Department': ['HR', 'Finance', 'IT', 'HR', 'Finance']
}
df = pd.DataFrame(data)

# Filtering rows where Salary > 60000
high_salary = df[df['Salary'] > 60000]
print("Filtered rows with Salary > 60000:\n", high_salary)

# Filtering rows where Department is 'Finance'
finance_dept = df[df['Department'] == 'Finance']
print("Filtered rows where Department is 'Finance':\n", finance_dept)

# Subsetting specific columns (Name and Salary)
subset_df = df[['Name', 'Salary']]
print("Subset of Name and Salary columns:\n", subset_df)

# Filtering based on multiple conditions (Salary > 60000 and Age < 45)
filtered_multiple = df[(df['Salary'] > 60000) & (df['Age'] < 45)]
print("Filtered rows with Salary > 60000 and Age < 45:\n", filtered_multiple)
```

```python
# Filtered data for visualization
filtered_data = df[df['Salary'] > 60000]

# Plotting filtered data
plt.bar(filtered_data['Name'], filtered_data['Salary'])
plt.title('Employees with Salary > 60000')
plt.xlabel('Employee Name')
plt.ylabel('Salary')
plt.show()
```
