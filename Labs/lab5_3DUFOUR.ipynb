## Lab 5.3: Network Visualization and Analysis

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community import greedy_modularity_communities
import numpy as np

graph = nx.Graph()
graph.add_edges_from([
    ('Alice', 'Bob'),
    ('Alice', 'Charlie'),
    ('Bob', 'Diana'),
    ('Charlie', 'Eve'),
    ('Eve', 'Ian'),
    ('Grace', 'Hannah'),
    ('Diana', 'Ian'),
    ('Ian', 'Grace'),
    ('Frank', 'Charlie'),
    ('Hannah', 'Jack')
])

nx.draw(graph, with_labels=True, node_color='green')
plt.show()
```

```python
# degree analysis

degree_centrality = nx.degree_centrality(graph)
print("Degree Analysis:")
for node, score in degree_centrality.items():
    print(f"{node}: {score}")
```

```python
# centrality measure
betweenness_centrality = nx.betweenness_centrality(graph)
print("Betweenness Centrality:")
for node, score in betweenness_centrality.items():
    print(f"{node}: {score}")
```

```python
# closeness centrality
closeness_centrality = nx.closeness_centrality(graph)
print("Closeness Centrality:")
for node, score in closeness_centrality.items():
    print(f"{node}: {score}")
```

```python
# community measure
communities = greedy_modularity_communities(graph)
for i, community in enumerate(communities, 1):
    print(f"Community {i}: {list(community)}")
```

```python
deg_centrality = nx.degree_centrality(graph)
centrality = np.fromiter(deg_centrality.values(), float)
nx.draw(graph, node_color=centrality, with_labels=True)
```
