#### What is the gender distribution across the organization?

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Emp_Data_Lab1.csv')

#print(data.head())

gender_counts = data['Gender'].value_counts()
gender_counts.plot(kind='pie', autopct='%1.1f%%', title='Gender Distribution')
```

#### What is the distribution of employees across different departments?

```python
# count jobs and create bar plot
job_level_counts = data['Job Level'].value_counts()
job_level_counts.plot(kind='bar',title='Job Level Distribution')
plt.xlabel('Job Level')
plt.ylabel('Number of Employees')
plt.show()
```

#### What is the distribution of employees across different departments?

```python
# count employees across departments and plot using pie chart
employees_per_dept_count = data['Department'].value_counts()
employees_per_dept_count.plot(kind='pie', autopct='%1.1f%%', title='Employees per Department')
```

#### How does salary vary with years of experience?

```python
# create scatter plot of salary * years of experience

# sorts the salary column before plotting it
#data.sort_values('Salary ($)', ascending=True)

# check data type to see if salary is numeric
#print(data.dtypes)

#converting salary from object to numeric
data['Salary ($)'] = data['Salary ($)'].str.replace(',','')
data['Salary ($)'] = pd.to_numeric(data['Salary ($)'])


data.plot.scatter(x='Years of Experience', y='Salary ($)', title='Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary ($)')
```

```python
# create histogram
data['Age'].plot.hist(bins=20, title='Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

#### Is there a correlation between height and salary?

```python
# create scatter plot of salary * years of experience

data.sort_values('Salary ($)', ascending=True, inplace=True)

data.plot.scatter(x='Height (cm)', y='Salary ($)', title='Salary vs. Height')
plt.xlabel('Height (cm)')
plt.ylabel('Salary ($)')
```
